```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

import keras
from keras import backend as K
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
img = load_img('/content/car.jpg')  # This is a PIL image
x = img_to_array(img)  # Convert to numpy array
x = x.reshape((1,) + x.shape)
plt.imshow(img)
plt.show()
x
```

```python
# Rotation augmentation
rotation_datagen = ImageDataGenerator(rotation_range=40)
rotation_datagen.fit(x)

i = 0
for img_batch in rotation_datagen.flow(x, batch_size=9, shuffle=False):
    plt.subplot(3, 3, i+1)
    plt.imshow(array_to_img(img_batch[0]))
    i += 1
    if i >= 9:
        break
plt.show()
```

```python
# Brightness augmentation
brightness_datagen = ImageDataGenerator(brightness_range=[0., 3.0])
brightness_datagen.fit(x)

i = 0
for img_batch in brightness_datagen.flow(x, batch_size=9, shuffle=False):
    plt.subplot(3, 3, i+1)
    plt.imshow(array_to_img(img_batch[0]))
    i += 1
    if i >= 9:
        break
plt.show()
```

```python
# Width , Height shift augmentation
width_height_datagen = ImageDataGenerator(width_shift_range=0.5, height_shift_range=0.3)
width_height_datagen.fit(x)

i = 0
for img_batch in width_height_datagen.flow(x, batch_size=9, shuffle=False):
    plt.subplot(3, 3, i+1)
    plt.imshow(array_to_img(img_batch[0]))
    i += 1
    if i >= 9:
        break
plt.show()
```

```python
# Zoom augmentation
zoom_datagen = ImageDataGenerator(zoom_range=0.7)
zoom_datagen.fit(x)

i = 0
for img_batch in zoom_datagen.flow(x, batch_size=9, shuffle=True):
    plt.subplot(3, 3, i+1)
    plt.imshow(array_to_img(img_batch[0]))
    i += 1
    if i >= 9:
        break
plt.show()
```

```python
# Horizontal flip augmentation
horizontal_flip_datagen = ImageDataGenerator(horizontal_flip=True)
horizontal_flip_datagen.fit(x)

i = 0
for img_batch in horizontal_flip_datagen.flow(x, batch_size=9, shuffle=False):
    plt.subplot(3, 3, i+1)
    plt.imshow(array_to_img(img_batch[0]))
    i += 1
    if i >= 9:
        break
plt.show()
```
